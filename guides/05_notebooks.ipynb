# Cosas √∫tiles que veremos en el curso üí°

Esto no es una caracter√≠stica de los notebooks en s√≠, pero hay un paquete muy √∫til llamado **`tqdm`** que muestra una **barra de progreso** al recorrer cualquier iterable.

````python
# Ejemplo sin barra de progreso (tardar√° 10 segundos y no sabr√°s qu√© pasa üòÖ)

import time

spams = ["spam"] * 1000

for spam in spams:
    time.sleep(0.01)
````

````python
# Ahora con barra de progreso usando tqdm üëá

import time
from tqdm import tqdm

spams = ["spam"] * 1000

for spam in tqdm(spams):
    time.sleep(0.01)
````
